Vou utilizar o método min-max nesse exercício. As colunas escolhidas para padronização são <b>Recognition Score</b> e <b>'Temporal Memory Score'</b>. Nossa fórmula é:

(x-min)/(max-min)

Onde x é o valor assumido pelo atributo.

```python
#importando libs
import pandas as pd
```

```python
#Carregando os dados
dados = pd.read_excel("./Dados/BBC-Results -MT10 Limpeza.xlsx")
dados.head(15)
```

```python
dados.min()
```

```python
dados.max()
```

Os valores máximo e mínimo para ambos os atributos foram 0 e 100, respectivamente. Logo nossa formula de min-max sofrerá as seguintes alteraçoes:
   
(x-min)/(max-min) => (x-0)/(100-0) => x/100

Então, apara fazermos a normalização das colunas, teremos apenas que dividir o valor por 100.

```python
#É importante que se execute esta linha apenas uma vez
dados['Recognition Score'] = dados['Recognition Score']/100
dados['Temporal Memory Score'] = dados['Temporal Memory Score']/100
dados.head(15)
```
