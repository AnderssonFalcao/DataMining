```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_bbc = pd.read_csv("BBC Results.csv")
```

#### Dropar tuplas que todos os registros sejam NAN

```python
df_bbc.count()
```

```python
df_bbc.dropna(axis=0, how='all')
```

#### Removendo colunas desnecessárias

```python
df_bbc = df_bbc.drop('Unnamed: 4', axis=1)
df_bbc = df_bbc.drop('Unnamed: 5', axis=1)
```

```python
df_bbc.head()
```

```python
mediana_sleephours = df_bbc['How many hours did you sleep last night?'].median()
```

#### Trocando valores da coluna que sejam NAN pela mediana

```python
df_bbc['How many hours did you sleep last night?'] = df_bbc['How many hours did you sleep last night?'].replace("NAN", mediana_sleephours)
```

#### Excluindo tuplas que tenham 50% dos campos com NAN

```python
df_bbc = df_bbc.dropna(thresh=3)  #thresh=3, mantém apenas tuplas que tenham no mínimo 3 campos != NAN
```

```python
df_bbc.count()
```

#### Excluindo o caracter % existentes em alguns valores  

```python
df_bbc['Recognition Score'] = df_bbc['Recognition Score'].replace("%", '')
```

```python
df_bbc
```

#### Analisando a completude dos dados, e substituindo os valores que não estão dentro do padrão desejado

```python
df_bbc = df_bbc.replace("ee", pd.np.nan)
```

```python
df_bbc
```

#### Apagando tuplas que tenham mais de 50% dos campos com NaN

```python
df_bbc = df_bbc.dropna(thresh=3)
```
