# Crawler site bitcointalk.org

```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("https://bitcointalk.org/index.php?board=67.0")

soup = BeautifulSoup(page.content, 'html.parser')
tudo = soup.findAll('td', class_='windowbg3')
tudo2 = soup.findAll('td', class_='windowbg')
```

```python
for tdo in tudo:
    print(tdo.text)
    
for tdo2 in tudo2:
    print(tdo2.text)
```
