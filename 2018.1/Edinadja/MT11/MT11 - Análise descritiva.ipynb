### 4. Na UFERSA Angicos gasta-se muito com energia elétrica. Observando os relatórios das despesas descentralizadas, é possível perceber um aumento de aproximadamente 28% entre 2015 e 2017 no gasto com energia elétrica. Quais as possíveis causas desse aumento e há alguma forma de mudar essa realidade? 

```python
import pandas as pd
import matplotlib.pyplot as plt

#Despesas de 2017
x = ['JAN','FEV','MAR','ABR','MAI','JUN','JUL','AGO','SET','OUT','NOV']
y = [19181.70, 35118.33, 40427.86,36626.66,41831.68,31673.67,38177.03,42899.09,45046.06,43557.83,34007.34]

#Despesas de 2015
x2 = ['JAN','FEV','MAR','ABR','MAI','JUN','JUL','AGO','SET','OUT','NOV']
y2 = [18819.35,35473.57,21461.21,28703.23,41099.32,33088.51,20739.58,19862.82,20265.81,19568.86,34449.11]

plt.plot([],[], color = 'g', label = 2017)
plt.plot([],[], color = 'r', label = 2015)
plt.plot(x, y, color = 'g')
plt.plot(x2, y2, color = 'r')
plt.title('Despesas com Energia - 2017 e 2015')
plt.legend()
plt.xlabel('Meses')
plt.ylabel('Gastos')

plt.show()
```

Como pode-se observar através dos gráficos apresentados acima, houve aumento no gasto com energia elétrica entre os anos de 2015 e 2017, e as possíveis causas deste aumento são: a ampliação do campus, com a construção de novos prédios; e o aumento na quantidade de ingressantes nos cursos, aumentando, assim, o consumo de energia nos blocos de salas de aula e laboratórios, o que tende a aumentar ainda mais com o passar do tempo.

Uma possível solução para diminuir os gastos com energia elétrica seria investir em placas de energia solar, que poderiam ser instaladas no campus, que se localiza em uma cidade bastante ensolarada a maior parte do ano.
