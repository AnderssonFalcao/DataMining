```python
import pandas as pd

df = pd.read_csv('BBC-Results.csv')
df
```

### Verificando o tipo de dados das colunas

```python
df.dtypes
```

### Ajustando o tipo de dados das colunas

```python
def ajusta_tipo_coluna(df, lista_colunas):
    df[lista_colunas] = df[lista_colunas].apply(pd.to_numeric, errors='coerce')

def ajusta_tipo_coluna2(df, lista_colunas2):
    df[lista_colunas2] = df[lista_colunas2].apply(pd.to_datetime, errors='coerce')
    
ajusta_tipo_coluna(df,['How many hours did you sleep last night?','Recognition Score','Temporal Memory Score'])
ajusta_tipo_coluna2(df,['Timestamp'])
```

```python
df
```

### Substitui valores nulos pela média:

```python
colunas = ['How many hours did you sleep last night?','Recognition Score','Temporal Memory Score']
for coluna in colunas:
    df[coluna].fillna(df[coluna].mean(), inplace = True)
df
```

### Substitui valores fora do limite pela média:

```python
colunas = ['How many hours did you sleep last night?','Recognition Score','Temporal Memory Score']
for coluna in colunas:
    df[(df[coluna] > 100) | (df[coluna] < 0)] = df[coluna].mean()
df
```

### Salva o arquivo limpo em .csv:

```python
df.to_csv('BBC-Results-MT9.csv', sep = ',', encoding='utf-8', index=False)
print ("Done!")
```
